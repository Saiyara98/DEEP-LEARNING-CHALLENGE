```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
nbic = application_df.drop(['EIN','NAME'], axis=1)
```

```python
uvoec = application_df.nunique()
uvoec
```

```python
atc = application_df['APPLICATION_TYPE'].value_counts()
atc
```

```python
coffvalve= 10
appliction_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts()<coffvalve].index.tolist()

for app in appliction_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

application_df['APPLICATION_TYPE'].value_counts()
```

```python
ctc = application_df['APPLICATION_TYPE'].value_counts()
ctc
```

```python
cvc = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts()>1]
cvc
```

```python
coffvalve_2 = 10
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts()<coffvalve_2].index.tolist()

for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

application_df['CLASSIFICATION'].value_counts()
```

```python
appdfe = pd.get_dummies(application_df)
appdfe.head(10)
```

```python
X = appdfe['IS_SUCCESSFUL']
y = appdfe.drop('IS_SUCCESSFUL', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
X_scaler = StandardScaler()
```

```python
X_train_scaled = X_scaler.fit_transform(X_train.values.reshape(-1, 1))
```

```python
X_test_scaled = X_scaler.transform(X_test.values.reshape(-1, 1))
```

```python
X_test_np = X_test.to_numpy()
X_test_np = X_test_np.reshape(-1, 1)
X_test_scaled = X_scaler.transform(X_test_np)
```

```python
inpu_feat= X_train_scaled.shape[1]
hidden_layer1_nodes = 80
hidden_layer2_nodes = 30
output_layer_nodes = 1

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=hidden_layer1_nodes, input_dim=inpu_feat, activation='relu'))
nn.add(tf.keras.layers.Dense(units=hidden_layer2_nodes,activation='relu'))
nn.add(tf.keras.layers.Dense(units=output_layer_nodes,activation='sigmoid'))

nn.summary()
```

```python
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train.iloc[:, 0])
ttm = nn.fit(X_train_scaled, y_train_encoded, validation_split=0.15, epochs=100)
```

```python
y_test_single_column = y_test['ASK_AMT']
y_test_np = y_test_single_column.to_numpy()
label_encoder = LabelEncoder()
y_test_encoded = label_encoder.fit_transform(y_test_np)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test_encoded, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
